```python
# Add anna to the path
import os
import sys
module_path = os.path.abspath(os.path.join('../../../anna'))
if module_path not in sys.path:
    sys.path.append(module_path)

DATA_DIR = "../../../data"
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import dataset.reuters21578.parser as data
from model.mlp import MLP

%matplotlib inline
```

```python
# Load data
train_docs, test_docs, unused_docs = data.fetch_and_parse(DATA_DIR)
```

```python
# Preprocess labels
labels = []
for d in train_docs + test_docs:
    for l in d.labels:
        if l not in labels:
            labels.append(l)
```

```python
# Create MLP classifier
model = MLP(DATA_DIR, labels, num_layers=2, voc_size=300000, hinge=True, verbose=True)
```

```python
# Train
history = model.train(train_docs, test_docs, epochs=200)
```

```python
pre, ax = plt.subplots(figsize=[15, 8])

x = range(len(history.history["loss"]))

ax.set_xlabel("Epoch")
ax.set_ylabel("Loss")
ax.plot(x, history.history["loss"], "b<", label="Train")
ax.plot(x, history.history["val_loss"], "r.", label="Validation")
ax.legend(loc=7)

pre, ax = plt.subplots(figsize=[15, 8])
ax.set_ylabel("Accuracy")
ax.plot(x, history.history["test_acc"], "g--", label="Test")
ax.plot(x, history.history["val_acc"], "r-", label="Validation")
ax.legend(loc=7)
```
